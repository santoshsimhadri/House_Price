```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
os.getcwd()
```

```python
os.chdir("Downloads/")
```

```python
data_set=pd.read_excel("DS - Assignment Part 1 data set.xlsx")
```

```python
data_set.head()
```

```python
data_set.isnull().sum()
```

```python
data_set.describe()
```

```python
x=data_set.iloc[:,1:8].values
x
```

```python
y=data_set.iloc[:,-1].values
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
x_train=StandardScaler().fit_transform(x_train)
x_train
```

```python
x_test=StandardScaler().fit_transform(x_test)
x_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model_LR=LinearRegression().fit(x_train,y_train)
```

```python
y_pred_LR=model_LR.predict(x_test)
```

```python
y_pred_LR
```

```python
from sklearn import metrics
```

```python
from sklearn.metrics import r2_score
```

```python
r2_LR=r2_score(y_test,y_pred_LR)
```

```python
print('r2 score for LinearRegression model is',r2_LR)
```

```python
from sklearn import metrics
print('Mean_absolute_error: ',metrics.mean_absolute_error(y_test,y_pred_LR))
print("Mean_squared_error: ",metrics.mean_squared_error(y_test,y_pred_LR))
print('Root_mean_squared_error: ',np.sqrt(metrics.mean_squared_error(y_test,y_pred_LR)))
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y_test,y_pred_LR)
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
```

```python
sns.regplot(x=y_test,y=y_pred_LR,ci=None,color ='red')
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
model_DT=DecisionTreeRegressor(criterion="poisson",random_state=0).fit(x_train,y_train)
```

```python
y_pred_DT=model_DT.predict(x_test)
```

```python
y_pred_DT
```

```python
r2_DT=r2_score(y_test,y_pred_DT)
```

```python
print('r2 score for DecisionTreeRegressor model is',r2_DT)
```

```python
print('Mean_absolute_error: ',metrics.mean_absolute_error(y_test,y_pred_DT))
print("Mean_squared_error: ",metrics.mean_squared_error(y_test,y_pred_DT))
print('Root_mean_squared_error: ',np.sqrt(metrics.mean_squared_error(y_test,y_pred_DT)))
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y_test,y_pred_DT)
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
```

```python
sns.regplot(x=y_test,y=y_pred_DT,ci=None,color ='green')
```

```python
model_RF=RandomForestRegressor(criterion="poisson").fit(x_train,y_train)
```

```python
y_pred_RF=model_RF.predict(x_test)
```

```python
y_pred_RF
```

```python
r2_RF=r2_score(y_test,y_pred_RF)
```

```python
print('r2 score for RandomForestRegressor model is',r2_RF)
```

```python
print('Mean_absolute_error: ',metrics.mean_absolute_error(y_test,y_pred_RF))
print("Mean_squared_error: ",metrics.mean_squared_error(y_test,y_pred_RF))
print('Root_mean_squared_error: ',np.sqrt(metrics.mean_squared_error(y_test,y_pred_RF)))
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y_test,y_pred_RF)
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
```

```python
sns.regplot(x=y_test,y=y_pred_RF,ci=None,color ='blue')
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y_test,y_pred_svm)
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
```
